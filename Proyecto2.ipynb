Descarga de la tabla "Tasa de Desocupación. Derecho al trabajo. Recepción del derecho" de la pagina de la INEGI.

Se crea la carpeta "data" en la cual se guarda un documento de texto con la informacion de la descarga y la informacion en formato csv.

```python
fechaactual = datetime.datetime.now().strftime("%d.%m.%Y_%H.%M")

Desocupacion= pd.read_csv('https://www.inegi.org.mx/contenidos/datosgob/tar02a_tasa_desocupacion.csv',encoding = 'ISO-8859-1')

os.makedirs("./data/Desocupacion_"+fechaactual, exist_ok=True)  
Desocupacion.to_csv("./data/Desocupacion_"+fechaactual+"/TasaDeDesocupación.DerechoAlTrabajo.RecepciónDelDerecho_"+fechaactual+".csv")  

file = open("./data/Desocupacion_"+fechaactual+"/DetallesDeDescarga"+fechaactual+".txt", "w")
file.write("Dentro de esta carpeta se encuentra el archivo csv= Tasa de Desocupación. Derecho al trabajo. Recepción del derecho"+fechaactual+".csv" + os.linesep)
file.write("Esta informacion fue descargada el dia ="+time.strftime("%x")+ os.linesep+"Hora"+time.strftime("%X")+ os.linesep)
file.write("enlace de los datos : http://www.cenatra.salud.gob.mx/transparencia/datos-abiertos/descarga/Pacientes_en_espera.xlsx" + os.linesep)
file.write("Pagina donde se encontro la informacio= https://www.datos.gob.mx/busca/dataset/tasa-de-desocupacion-derecho-al-trabajo-recepcion-del-derecho"+ os.linesep)
file.write("Publicado por: juan.nieto@inegi.org.mx ")
file.close()

UbiDes="./data/Desocupacion_"+fechaactual+"/TasaDeDesocupación.DerechoAlTrabajo.RecepciónDelDerecho_"+fechaactual+".csv"
desocupacion=pd.read_csv(UbiDes)#,encoding = 'ISO-8859-1')
desocupacion.head(1)
```

```python
fechaactual = datetime.datetime.now().strftime("%d.%m.%Y_%H.%M")

Duracion= pd.read_csv('https://www.inegi.org.mx/contenidos/datosgob/tcr02_duracion_desocupacion.csv',encoding = 'ISO-8859-1')

os.makedirs("./data/duracion_"+fechaactual, exist_ok=True)  
Duracion.to_csv("./data/duracion_"+fechaactual+"/Duraciondeladesocupacion"+fechaactual+".csv")  

file = open("./data/duracion_"+fechaactual+"/DetallesDeDescarga"+fechaactual+".txt", "w")
file.write("Dentro de esta carpeta se encuentra el archivo csv= Duración de la desocupación. Derecho al trabajo. Capacidades estatales"+fechaactual+".csv" + os.linesep)
file.write("Esta informacion fue descargada el dia ="+time.strftime("%x")+ os.linesep+"Hora"+time.strftime("%X")+ os.linesep)
file.write("enlace de los datos : https://www.inegi.org.mx/contenidos/datosgob/tcr02_duracion_desocupacion.csv" + os.linesep)
file.write("Pagina donde se encontro la informacio= https://datos.gob.mx/busca/dataset/duracion-de-la-desocupacion-derecho-al-trabajo-capacidades-estatales1"+ os.linesep)
file.close()

Ubidur="./data/duracion_"+fechaactual+"/Duraciondeladesocupacion"+fechaactual+".csv"
duracion=pd.read_csv(Ubidur)#,encoding = 'ISO-8859-1')
duracion.head(1)
```
